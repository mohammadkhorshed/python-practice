```python
# Exercise 1: Write a function `count_vowels(s)` that returns the number of vowels in a string.

def count_vowels(sent):
    return sum(1 for c in sent.lower() if c in 'aeiou')

count_vowels("I am a Hero!")
```

```python
def count_vowel(s):
    count = 0
    for i in s.lower():
        if i in 'aeiou':
            count+=1
    return count
count_vowel("I am a Hero!")
```

```python
# Exercise 2: Write a function `is_palindrome(s)` that checks if a string is a palindrome.

def is_palindrome(s):
        if s.lower().strip() == s[::-1].lower().strip():
                return "This string is a palindrome!"
        else:
                return "This string is not a palindrome!"

is_palindrome("Radar")
```

```python

def is_palindrome(s):       # single-liner
    return "This string is a palindrome!" if s.lower().strip() == s[::-1].lower().strip() else "This string is not a palindrome!"

is_palindrome("Madam")
```

```python
# Exercise 3: Using a single line, remove all vowels from the string `'Beautiful day in the neighborhood'`.

def remove_vowel(s):
    return ''.join(c for c in s if c.lower() not in 'aeiou')
remove_vowel('Beautiful day in the neighborhood')
```

```python
def remove_vowel(s):
    for l in s.lower():
        if l in 'aeiou':
           s = s.replace(l, '')
    return s

remove_vowel("Beautiful day in the neighborhood")
```

```python
# Exercise 4: Extract the domain from an email address (e.g., `'john.doe@gmail.com'` → `'gmail.com'`).

def domain_extractor(d):
    return d.split("@")[-1]
domain_extractor('john.doe@gmail.com')
```
