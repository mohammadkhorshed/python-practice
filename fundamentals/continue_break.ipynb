### 1. Print all numbers from 1 to 10 except 5 using `continue`.

```python
for num in range(1, 11):
    if num == 5:
        continue
    print(num, end=' ')
```

### 2. Print numbers from 1 to 20, but stop the loop if the number is divisible by 7 using `break`.

```python
for num in range(1, 21):
    if num % 7 == 0:
        break
    print(num, end=' ')
```

### 3. Loop through a list of numbers and skip all even numbers using `continue`.

```python
lst = [13, 17, 6, 21, 14, 11, 5, 8, 15]

for n in lst:
    if n % 2 == 0:
        continue
    print(n, end=' ')
```

### 4. From a list of names, stop printing names once you encounter the name 'Stop' using `break`.

```python
names = ['Atif', 'Aslam', 'Sonu', 'Nigam', 'Stop', 'Hasan']

for name in names:
    if name == 'Stop':
        break
    print(name)
```

### 5. Print the first 5 non-negative numbers divisible by 3, skipping number 0 using `continue`.

```python
num, count = 0, 0
while count < 5:
    if num == 0:
        num += 1
        continue
    if num % 3 == 0:
        print(num, end = ' ')
        count += 1
    num += 1
```

### 6. Loop through a sentence and print all characters until a space is encountered using `break`.

```python
text = "Data is the new oil."

for char in text:
    if char == ' ':
        break
    print(char, end=' ')
```

### 7. Given a list of numbers, print the squares but skip negative numbers using `continue`.

```python
lst = [2, -4, 7, 9, -2, -9, 8, 5, -3, 6]

for n in lst:
    if n < 0:
        continue
    print(n ** 2, end=' ')
```

### 8. Create a login simulation where the user gets 3 tries, but the loop `break`s after a correct password.

```python
tries = 3
password = 'Python123'
while tries > 0:
    input_password = input('Enter password: ')
    if input_password == password:
        print('Access granted!')
        break
    tries -= 1
    print(f"Attempt remaining: {tries}")
    if tries == 0:
        print('Blocked! Contact admin to reset your password!')
```

### 9. Write a loop that prints the first 10 prime numbers using `break` and `continue` where needed.

```python
count, num = 0, 2
while count < 10:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if not is_prime:
        num += 1
        continue
    print(num, end=' ')
    count += 1
    num += 1
```

### 10. From a nested loop (2D list), break out of the **inner loop** when a zero is encountered and continue outer loop.

```python
mtx = [
    [1, 2, 3],
    [4, 0, 5],
    [6, 7, 8]
]

for row in mtx:
    for num in row:
        if num == 0:
            print("[Incomplete row]")
            break
        print(num, end=' ')
    else:
        print()
    
```
