### 18. Write a function that accepts a tuple of numbers and returns the sum.

```python

def sum_of_num(tpl):
    total = 0
    for num in tpl:
        total += num
    return total

my_tpl = (25, 12, 54, 32)
sum_of_num(my_tpl)
```

### 19. From a tuple of strings, return the longest one.

```python
# Solution-1
texts = ("Bangladesh", "United Sates", "Germany", "United Kingdom", "Canada", "United Arab Emirates")

longest_str = ""

for text in texts:
    if len(text) > len(longest_str):
        longest_str = text
print(longest_str)
```

```python
# Solution-2
texts = ("Bangladesh", "United Sates", "Germany", "United Kingdom", "Canada", "United Arab Emirates")

longest_str = max(texts, key=len)
print(longest_str)
```

### 20. Given a tuple of (name, age) pairs, return a list of names where age > 25.

```python
people = (("Mohammad", 25), ("John", 36), ("Rakib", 22), ("Ahmed", 26), ("Rajib", 33))

name_list = [person[0] for person in people if person[1] > 25]

print(name_list)
```
