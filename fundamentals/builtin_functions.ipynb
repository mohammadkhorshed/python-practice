### 1. Practice using `abs()`

```python
num = -11
abs_num = abs(num)
print(abs_num)
```

```python
img_num = 3 + 4j
abs_img_num = abs(img_num)
print(abs_img_num)
```

### 2. Practice using `round()`

```python
pi = 3.1415926535
rounded_pi = round(pi, 2)
print(rounded_pi)
```

### 3. Practice using `divmod()`

```python
x = 10
y = 3
result = divmod(x, y)
print(result)
```

### 4. Practice using `pow()`

```python
a = 5
b = 3
result = pow(a, b)
print(result)
```

### 5. Practice using `int()`

```python
n = 12.2
int_n = int(n)
print(int_n)
```

### 6. Practice using `float()`

```python
float_n = float(int_n)
print(float_n)
```

### 7. Practice using `complex()`

```python
num = 7
complex_num = complex(num)
print(complex_num)
```

### 8. Practice using `bin()`

```python
num = 10
binary_num = bin(num)
print(binary_num)
```

### 9. Practice using `hex()`

```python
num = 45
hex_num = hex(num)
print(hex_num)
```

### 10. Practice using `oct()`

```python
num = 15
oct_num = oct(num)
print(oct_num)
```
