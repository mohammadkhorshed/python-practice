### 14. Given an array, find and return all elements greater than the average.

```python
def greater_than_avg(ary):
    return [n for n in ary if n > (sum(ary) / len(ary))]

greater_than_avg(ar3)
```

### 17. Create a function that removes all duplicates from an array.

```python
# solution-1: Using set, doesn't maintain order
def remove_duplicates(ary):
    unique_set = set(ary)
    return ar.array(ary.typecode, unique_set)

ar4 = ar.array("i", [2, 4, 3, 7, 2, 3, 7, 1])
remove_duplicates(ar4)
```

```python
# Solution-2: Maintains order
def remove_duplicates(ary):
    unique_array = ar.array(ary.typecode, [])
    for n in ary:
        if n not in unique_array:
            unique_array.append(n)
    return unique_array

remove_duplicates(ar4)
```
