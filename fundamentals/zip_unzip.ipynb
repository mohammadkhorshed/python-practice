### 1. Use `zip()` to combine two lists: `['a', 'b', 'c']` and `[1, 2, 3]`.

```python
lst1 = ['a', 'b', 'c']
lst2 = [1, 2, 3]

ziped = list(zip(lst1, lst2))
print(ziped)
```

### 2. Create a dictionary from two lists using `zip()`: keys = ['name', 'age'], values = ['John', 25].

```python
keys = ['name', 'age']
values = ['John', 25]

my_dict = dict(zip(keys, values))
print(my_dict)
```

### 3. Zip three lists together: names, ages, and cities.

```python
names = ['John', 'Mehmed', 'Uthman']
ages = [36, 21, 48]
cities = ['Bethany', 'Constantinople', 'Hejaz']

zipped_list = list(zip(names, ages, cities))
print(zipped_list)
```

### 4. Unzip a list of tuples into individual lists using `zip(*zipped)`.

```python
unzipped_names, unzipped_ages, unzipped_cities = map(list, zip(*zipped_list))

print(unzipped_names)
print(unzipped_ages)
print(unzipped_cities)
```

### 5. Combine first and last names lists into full names using `zip()` and list comprehension.

```python
first_names = ['Sakib', 'Salman', 'Johnny', 'Al', 'Tim']
lst_names = ['Khan', 'Shah', 'Depp', 'Pacino', 'Robbins']

full_name = [f"{first.strip().title()} {last.strip().title()}" for first, last in zip(first_names, lst_names)]
print(full_name)
```

### 6. Zip a list of student names with their marks and print only those with marks above 80.

```python
students = ['Sakib Khan', 'Salman Shah', 'Johnny Depp', 'Al Pacino', 'Tim Robbins']
marks = [75, 94, 80, 98, 86]

student_list = list(zip(students, marks))
print([student for student in student_list if student[1] > 80])
```
