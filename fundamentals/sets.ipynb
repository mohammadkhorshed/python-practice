### 21. Write a function that returns the intersection of all sets in a list.

```python
# Solution-1: Solved with my own effort 
def find_intersection(lst):
    if not lst:
        return set()
    for i in range(len(lst)-1):
        lst[i+1].intersection_update(lst[i])
    return lst[-1]

find_intersection([{1,2,3}, {2,3,4}, {3,4,5}])
    
```

```python
# Solution-2: Solved with help of DeepSeek
def find_intersection(lst):
    if not lst:
        return set()
    result_set = lst[0]
    for ele in lst[1:]:
        result_set.intersection_update(ele)
    return result_set

find_intersection([{1,2,3}, {2,3,4}, {3,4,5}])
```

### 22. Given a list of strings, return all unique words.

### 23. Find the first non-repeating character in a string using set logic.

```python
text = "data science"

def first_non_repeating_char(string):
    freq = {}
    seen = set()
    candidate = []
    for char in string:
        if char not in seen:
            seen.add(char)
            candidate.append(char)
        freq[char] = freq.get(char, 0) + 1
    for char in candidate:
        if freq[char] == 1:
            return char
    return None

first_non_repeating_char(text)
```

### 24. Write a program to find duplicate elements in a list using sets.

```python
def find_duplicates(lst):
    seen = set()
    duplicates = set()

    for ele in lst:
        if ele not in seen:
            seen.add(ele)
        else:
            duplicates.add(ele)
    return duplicates

my_list = [2, 4, 3, 7, 4, 2, 6]

find_duplicates(my_list)
    
```

### 25. You are given two datasets of users who bought product A and B. Find users who bought both, only A, and only B.

```python
# Solution-1
buyer_of_product_a = {"Asif", "Adnan", "Fahad", 'Russel', "Yash"}
buyer_of_product_b = {"Ayan", "Asif", "John", "Kabir", "Fahad"}

buyer_of_both = set()
buyer_of_only_a = set()
buyer_of_only_b = set()

for ele in buyer_of_product_a:
    if ele in buyer_of_product_b:
        buyer_of_both.add(ele)
    else:
        buyer_of_only_a.add(ele)
buyer_of_only_b = buyer_of_product_b.difference(buyer_of_both)

print(f"{buyer_of_both} brought both products.")
print(f"{buyer_of_only_a} brought only product A.")
print(f"{buyer_of_only_b} brought only product B.")
```
